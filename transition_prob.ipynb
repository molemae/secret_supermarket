# Difining the process as a function

```python
def shifting_location(df):
    '''This function making a new column in our dataframe with shifting the column 'location' to up'''
    df['location2'] = df['location'].shift(-1)
    return df
```

```python
def cleaning_location2(df):
    '''The location2 needs some cleaning, replacing entrance with chechout. '''
    df['location2'].replace(to_replace='entrance', value='checkout', inplace=True)
    df['location2'].replace(to_replace= np.nan, value='checkout', inplace=True)
    return df

```

```python
def transition_prob_matrix(df):
    mat = pd.crosstab(df['location'], df['location2'], normalize=0)
    return mat
```

```python
def trans_prob_mat_calculating(df):
    df_shift = shifting_location(df)
    df_clean = cleaning_location2(df_shift)
    trans_mat = transition_prob_matrix(df_clean)
    return trans_mat
```

# Another way to write the function

```python
def transition_probbability_matrix(dataframe):
    df = dataframe

    '''This line making a new column in our dataframe with shifting the column 'location' to up'''
    df['location2'] = df['location'].shift(-1)

    '''The location2 needs some cleaning, replacing entrance with chechout. '''
    df['location2'].replace(to_replace='entrance', value='checkout', inplace=True)
    df['location2'].replace(to_replace= np.nan, value='checkout', inplace=True)

    '''Here we creat the transition matrix (probability matrix)'''
    mat = pd.crosstab(df['location'], df['location2'], normalize=0)
    return mat
    
```
